```python
import pandas as pd    #imports the pandas library
```

```python
#Load the dataset from the URL using read_csv.

url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
data = pd.read_csv(url) #reads the csv file from the url given
data.head() #by default displays first 5 rows
#Reflection: I learned how to load a dataset directly from a URL and view first 5 rows using pandas
```

Reflection: learned to uplaod the dataset from a url using read_csv() and view the data using data.head()

```python
#Display the first 10 rows using head().

data.head(10)  #gives the number of rows mentioned in the paranthesis starting from zero
#Reflection: learned how to view fisrt few rows of a dataset
```

```python
#Use info() to see column types and non-null counts.

data.info()  #helps viewing datatype
#Reflection: I learned to check the data types and dataset information
```

```python
#Use describe() to get numeric summaries (mean, std, etc.).

data.describe()  #gives statistical summary of numerical columns
#Reflection: learned how to view the summary of numerical data like count, mean, standard deviation, minimum, maximum
```

```python
#Use isnull().sum() to check for missing values.

data.isnull().sum()   #detects and counts missing values and sum() adds the numbers in the table
#Reflection: I learned to count missing values in each column
```

```python
#Use nunique() to find how many unique Pokmon names, types, and generations there are.

data.nunique()  #counts number of unique or distinct values in specified axis
#reflection: I learned how to count the number of unique values
```

```python
#Use loc to select only those Pokmon where Type 1 is Water.

data.loc[data['Type 1'] == 'Water'] #selects data using row and column labels
#reflection: shows how to select rows based on a condition
```

```python
#Use iloc to display rows 1020 and columns 14.

data.iloc[10:21, 1:5] #selects data by integer position
#reflection: this tells how to select data by their integer positions
```

```python
#Use sort_values() to find the top 5 Pokmon by Attack power.

data.sort_values(by='Attack', ascending=False).head(5)  #arrange data in order by column
#reflection: this shows how to sort the data by attack column in descending order and display the top 5 rows
```

```python
#Use query() to show Pokmon with Attack > 100 and Defense > 100.

data.query('Attack>100 and Defense>100') #filter columns using a condition
#reflection: this shows how to filter columns on multiple conditions
```

```python
#Use between() to find Pokmon whose Speed is between 50 and 100.

data[data['Speed'].between(50, 100)]   #checks if values falls within the specified range
#reflection: this shows how to filter rows where values fall within a specific range
```

```python
#Use groupby() to find the average HP by Type 1.

data.groupby('Type 1')['HP'].mean()   #used to group data and compute operations and mean() calculates the average of a value
#reflection: this shows how to group data by type1 and calculate average HP
```

```python
#Use agg() to find both min and max Attack for each Type 1.

data.groupby('Type 1')['Attack'].agg('min', 'max')  #use to apply one or more operations after grouping
#reflection: I learned how to apply multiple functions to grouped data using groupby() and agg()
```

```python
#Use pivot_table() to show average Defense by Type 1 and Generation.

pd.pivot_table(data, values='Defense', index=['Type 1', 'Generation'], aggfunc={'Defense':"mean"})
#pivot_table() creates table showing summaries of data by one or more categories
#aggfunc() specifies which calculation to do
#reflection: I learned how to create pivot table to summarize defense by type1 and generation using aggfunc to calculate mean
```

```python
#Use value_counts() to see which Type 1 appears most frequently.

data['Type 1'].value_counts()   #return a series of each unique value
#reflection: this shows how to get the count of each unique value using value_counts()
```

```python
#Use fillna() to replace missing Type 2 values with None.

data['Type 2'].fillna('None').head(15) #fills NA/NaN values using the specified method.
#reflection: i learned how to replace missing value in the type2 column with 'none' using fillna()
```

```python
#Use dropna() to remove rows with any missing data (and note how many rows remain).

print(data.dropna().shape) #Remove missing values and shape() return the the dimensionality of the data.
#reflection: this shows how to remove rows with missing values using dropna() and check the data's dimensions with shape
```

```python
#Use replace() to change all Fire in Type 1 to  Fire.

data.replace(regex='Fire', value= ' Fire') #can modify specific values
#reflection: this shows how to replace values using replace() with a regular expression
```

```python
#Create a function that categorizes Attack as Low, Medium, High and apply it to create a new column Attack_Level.

def categorize_attack(attack):
  if attack < 50:
     return('low')
  elif attack < 100:
     return('medium')
  else:
     return('high')

data['Attack_Level']=data['Attack'].apply(categorize_attack)   #data['Attack_Level']- creates new column, apply(categorize_attack)- categorize an
print(data[['Name', 'Attack', 'Attack_Level']])                 #existing column
#reflection: i learned to create  a new column by applying a function to categorize values  from an existing column using apply()
```

```python
#Use astype() to convert Generation into a string type.

data['Generation'].astype(str)  #convert a data from one type to another
#reflection: this shows how to convert the data type using astype()
```

```python
#Split your dataset into two halves and combine them using concat().

df1 = pd.DataFrame(data.iloc[:400, :])
df2 = pd.DataFrame(data.iloc[400:800, :])
combined = pd.concat([df1, df2])  #combines two or more datas either rows or columns
#reflection: this shows how to split a data into parts and then combine using concat()
```

```python
#Use duplicated() to check for duplicates.

data.duplicated() #identifies duplicate values
#reflection: this shows how to identify duplicate rows in data using duplicated()
```

```python
#Remove any duplicates using drop_duplicates().

data.drop_duplicates() #Return data with duplicate rows removed
#reflection: shows to remove duplicate rows from a data
```

```python
#Use melt() to reshape the dataset  show Name and the stats (HP, Attack, Defense, Speed) in long format.

pd.melt(data, id_vars=['Name'], value_vars=['HP', 'Attack', 'Defense', 'Speed'],
        var_name= 'Stat', value_name= 'Value') #Unpivot a DataFrame from wide to long format by turning columns into rows
        #reflection: I learned to reshape the data from wide to long format using melt()
```

```python
#Create a mini DataFrame with Pokmon names and ranks, then merge() it with your main DataFrame.

mini_df = pd.DataFrame({'Name': ['Bulbasaur', 'Ivysaur', 'Charmander', 'Diancie', 'Volcanion'],
                       'Rank': [1, 2, 3, 4, 5]})
merged = pd.merge(data, mini_df, on='Name', how='left')
print(merged)  #combines two datas by matching rows
#reflection: I learned to combine two datasets based on a common column using merge()
```

```python
#Use sample() to pick 8 random Pokmon.

data['Type 1'].sample(n=10) #Return a random sample of items of a data
#reflection: this shows how to select random sample of rows from a column using sample()
```

```python
#Use corr() to check which two attributes (among HP, Attack, Defense, Speed) are most correlated.

data[['HP', 'Attack', 'Defense', 'Speed']].corr() #shows how two or more columns are related to each other,  excluding NA/null values
#reflection: this shows how to find the correlation between numeric columns using corr()
```

```python
#Export your cleaned and modified DataFrame to a CSV file named pokemon_cleaned.csv using to_csv().

data.to_csv('pokemon_cleaned.csv') #saves a data as a CSV file
#reflection: i learned to save a dataset to a CSV file using to_csv()
```
